```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

iris = load_iris()
x = iris.data
y = iris.target.reshape(-1,1)
print(x)
print(y)

encoder = OneHotEncoder(sparse_output=False)
y_encoded = encoder.fit_transform(y)
print(y_encoded)

x_tr, x_ts, y_tr, y_ts = train_test_split(x,y_encoded,test_size=0.2,random_state=42)
x_tr_nn,x_ts_nn,y_tr_nn,y_ts_nn = train_test_split(x,y_encoded, test_size=0.2, random_state=42) ## nn "neural network"

# Machine Learning Approach, Decision Tree Classifier
tree_model = DecisionTreeClassifier()
tree_model.fit(x_tr,y_tr)
y_pred_tree = tree_model.predict(x_ts)
tree_accuracy = accuracy_score(y_ts, y_pred_tree)
print("Decision Tree Accuracy (ML):", tree_accuracy)
# Deep Learning Approach, Neural Network
nn_model = Sequential([
    Dense(16, activation="relu", input_shape=(4,)),
    Dense(16, activation="relu"),
    Dense(3,  activation="softmax")
])

nn_model.compile(optimizer="adam",
                 loss="categorical_crossentropy",
                 metrics=["accuracy"])

nn_model.fit(x_tr_nn,y_tr_nn,epochs=50,batch_size=5,validation_split=0.1,verbose=1)

nn_lose, nn_accuracy = nn_model.evaluate(x_ts_nn, y_ts_nn, verbose=1)

print("Neural Network Accuracy (DL):", nn_accuracy)

#Compare
print("Comparison of ML and DL Approaches:")
print("Decision Tree Accuracy (ML):", tree_accuracy)
print("Neural Network Accuracy (DL):", nn_accuracy)
```

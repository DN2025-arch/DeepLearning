```python
import numpy as np
from PIL import Image, ImageDraw
import tkinter as tk
from tensorflow.keras.models import load_model
```

```python
model = load_model(r"Zooms\DeepLearning\NewDigitRecognizerModel.h5")
```

```python
class DigitRecognizerApp:
    def __init__(self,model):
        self.model = model
        self.window = tk.Tk()
        self.window.title("Handwritten Digit Recognizer")

        self.canvas = tk.Canvas(self.window, width=200, height=200, bg="white")
        self.canvas.grid(row=0, column=0, pady=2, padx=2)
        self.canvas.bind("<B1-Motion>", self.draw)

        self.label = tk.Label(self.window, text="Draw A Digit", font=("Arial", 24))
        self.label.grid(row=0,column=1,padx=2,pady=2)

        self.clear_button = tk.Button(self.window,text="Clear",command=self.clear_canvas)
        self.clear_button.grid(row=1,column=0,padx=2,pady=2)

        self.predict_button = tk.Button(self.window,text="Predict")
        self.predict_button.grid(row=1,column=1,padx=2,pady=2)

        self.image = Image.new("L", (200,200), color="white")
        self.draw_obj = ImageDraw.Draw(self.image)

    def draw(self,event):
        x,y = event.x, event.y
        self.canvas.create_oval(x-8,y-8,x+8,y+8, fill="black",outline="black")
        self.draw_obj.ellipse([x-8,y-8,x+8,y+8],fill=0)

    def clear_canvas(self):
        self.canvas.delete("all")
        self.image = Image.new("L", (200,200), color="white")
        self.draw_obj = ImageDraw.Draw(self.image)
        self.label.config(text="Draw A Digit")

    def predict_digit(self):
        img = self.image.resize((28,28))
        img = np.array(img) / 255
        img = np.expand_dims(img,axis=0)
        prediction = self.model.predict(img)
        digit = np.argmax(prediction)
        self.label.config(text=f"Prediction: {digit}")

    def run(self):
        self.window.mainloop()

app = DigitRecognizerApp(model)
app.run()
```
